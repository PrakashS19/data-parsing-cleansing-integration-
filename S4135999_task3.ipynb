## Introduction
This task focuses on integrating the cleaned dataset from Task 2 with the provided dataset2.csv file. The main challenge was that both datasets followed different schemas, even though they contained similar information.

To solve this, the schema differences were identified and resolved by renaming columns and aligning their structure. After that, both datasets were combined into a single table.

Finally, the integrated dataset was checked for duplicate records and consistency, and a suitable global key approach was discussed based on the nature of the data.

##  Import libraries 

```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
from pathlib import Path
```

```python
base_path = Path(".")
dataset1_path = base_path / "S4135999_dataset1_solution.csv"
dataset2_path = base_path / "S4135999_dataset2.csv"
```

### 1. Examining and loading data
In this step, both datasets were loaded and examined to understand their structure.

The cleaned dataset from Task 2 contains review data extracted from XML, while dataset2.csv is already structured in a tabular format. By comparing their columns, it was observed that both datasets contain similar attributes but use different naming conventions.

This inspection helps identify schema conflicts before integration.

```python
dataset1 = pd.read_csv(dataset1_path)
dataset2 = pd.read_csv(dataset2_path)
```

```python
print("Dataset 1 shape:", dataset1.shape)
print("Dataset 2 shape:", dataset2.shape)
```

```python
print("Dataset 1 columns:")
print(dataset1.columns.tolist())

print("\nDataset 2 columns:")
print(dataset2.columns.tolist())
```

```python
dataset1.head()
```

```python
dataset2.head()
```

### 2. Resolving schema conflicts
Before merging the datasets, schema conflicts need to be resolved. Although both datasets contain similar information, their column names and structures are different.

The conflicts mainly include naming differences, column ordering, and data type inconsistencies. These issues must be fixed so both datasets follow the same schema.

### Conflict 1: Naming conflicts
The cleaned dataset uses source-based column names such as "ClothID", "Review Title", and "Customer Rating", while dataset2 uses standardised names like "Clothing ID", "Title", and "Rating".

Since these columns represent the same information, they were renamed to match the global schema used in dataset2.

```python
rename_map = {
    "ClothID": "Clothing ID",
    "Review Title": "Title",
    "Customer Rating": "Rating",
    "Positive Review Count": "Positive Feedback Count",
    "Section": "Division Name",
    "Department": "Department Name",
    "Category": "Class Name",
    "Online Time": "Active Time",
    "Cost": "Price"
}

dataset1_renamed = dataset1.rename(columns=rename_map)
dataset1_renamed.head()
```

### Conflict 2: Column order conflict
After renaming the columns, the order of columns in both datasets was still different.

To ensure consistency, both datasets were rearranged to follow the same column order based on the required global schema.

```python
global_schema = [
    "Clothing ID",
    "Age",
    "Title",
    "Rating",
    "Recommended IND",
    "Positive Feedback Count",
    "Division Name",
    "Department Name",
    "Class Name",
    "Active Time",
    "Price"
]
```

```python
dataset1_aligned = dataset1_renamed[global_schema]
dataset2_aligned = dataset2[global_schema]
```

```python
print("Dataset 1 aligned columns:")
print(dataset1_aligned.columns.tolist())

print("\nDataset 2 aligned columns:")
print(dataset2_aligned.columns.tolist())
```

### Conflict 3: Data type consistency
Even after aligning column names and order, the data types needed to be consistent across both datasets.

Numeric columns such as ID, rating, and price were converted to appropriate numeric types to ensure compatibility during merging.

```python
print("Dataset 1 data types:")
print(dataset1_aligned.dtypes)

print("\nDataset 2 data types:")
print(dataset2_aligned.dtypes)
```

```python
int_cols = [
    "Clothing ID",
    "Age",
    "Rating",
    "Recommended IND",
    "Positive Feedback Count"
]

float_cols = [
    "Active Time",
    "Price"
]

for col in int_cols:
    dataset1_aligned[col] = pd.to_numeric(dataset1_aligned[col], errors="coerce").astype("Int64")
    dataset2_aligned[col] = pd.to_numeric(dataset2_aligned[col], errors="coerce").astype("Int64")

for col in float_cols:
    dataset1_aligned[col] = pd.to_numeric(dataset1_aligned[col], errors="coerce")
    dataset2_aligned[col] = pd.to_numeric(dataset2_aligned[col], errors="coerce")
```

```python
print("Dataset 1 updated data types:")
print(dataset1_aligned.dtypes)

print("\nDataset 2 updated data types:")
print(dataset2_aligned.dtypes)
```

### 3. Merging data
After resolving all schema conflicts, the two datasets were merged into one integrated dataset.

Since both datasets now share the same structure, vertical merging using concatenation was used. This combines all records from both datasets into a single table.

```python
integrated_df = pd.concat([dataset1_aligned, dataset2_aligned], ignore_index=True)
integrated_df.head()
```

```python
print("Integrated dataset shape:", integrated_df.shape)
```

### 4. Resolving data conflicts:
After merging the datasets, the integrated data was checked for data-level conflicts.

The main issue identified was duplicate records. These duplicates can occur when the same record appears in both datasets. Removing them ensures that the dataset contains only unique entries.

### Conflict 1: Duplicate records
Exact duplicate rows were identified where all values in the row are the same.

These duplicates were removed to avoid redundancy and maintain data quality in the final dataset.

```python
exact_duplicates = integrated_df.duplicated()
print("Number of exact duplicate rows:", exact_duplicates.sum())
```

```python
integrated_df[exact_duplicates].head()
```

```python
integrated_df = integrated_df.drop_duplicates().reset_index(drop=True)
print("Shape after removing exact duplicates:", integrated_df.shape)
```

### Conflict 2: Repeated Clothing ID values
It was observed that the same Clothing ID can appear multiple times. This is expected because multiple customers can review the same clothing item.

Therefore, repeated Clothing ID values were not treated as errors. Only exact duplicate rows were removed.

```python
print("Number of unique Clothing ID values:", integrated_df["Clothing ID"].nunique())
print("Total number of rows:", len(integrated_df))
```

```python
integrated_df["Clothing ID"].value_counts().head(10)
```

#### Finding global key for the data
The dataset represents customer reviews rather than unique products. Therefore, a single column such as "Clothing ID" cannot act as a global key because it can have multiple associated reviews.

Instead, the uniqueness of the dataset is defined by the entire record. A composite key based on all columns can represent a unique review.

For this reason, duplicate handling was performed at the full-row level rather than relying on a single key column.

```python
print("Duplicate rows after cleaning:", integrated_df.duplicated().sum())
```

```python
candidate_key_check = integrated_df.duplicated(subset=global_schema).sum()
print("Duplicate count using all columns as composite key:", candidate_key_check)
```

### Handling Missing Values After Integration

After merging the datasets, some missing values were found in a few columns such as Title, Division Name, Department Name, and Class Name.

These missing values were not removed or replaced because they are very small compared to the total dataset size. Also, columns like Title are not critical for most analysis.

Removing rows would reduce the dataset unnecessarily, so the data was kept as it is after confirming that the missing values do not significantly affect the overall quality of the dataset.

After integration, additional validation checks were performed to ensure data quality.

First, summary statistics were examined to confirm that numeric values such as Age, Rating, and Price fall within reasonable ranges.

Second, categorical columns such as Division Name, Department Name, and Class Name were reviewed to ensure consistency and absence of unexpected values.

Finally, missing values were assessed and found to be minimal relative to the dataset size. Since they do not significantly affect analysis, no imputation or removal was performed.

These validation steps confirm that the integrated dataset is reliable and suitable for further analysis.
